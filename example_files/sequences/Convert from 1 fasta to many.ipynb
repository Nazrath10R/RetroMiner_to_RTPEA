```python
file1  = open("LINE_1.fasta", "r")
```

```python
file1_list = file1.readlines()
```

```python
file2 = ''.join(file1_list)
```

```python
file2
```

```python
#coding=utf8
# the above tag defines encoding for this document and is for Python 2.x compatibility

import re

regex = r">\D*\d*\D*\d*\D*\d*\D*\dp"

list_of_names=[]
test_str = (file2)

matches = re.finditer(regex, test_str, re.MULTILINE)

for matchNum, match in enumerate(matches):
    matchNum = matchNum + 1
    
#     print ("{match}".format(match = match.group()[1:]))
    list_of_names.append(match.group()[1:])
```

```python
str(list_of_names)
```

```python
#coding=utf8
# the above tag defines encoding for this document and is for Python 2.x compatibility

import re

regex = r">\D*\d*\D*\d*\D*\d*\D*\d*\D*$"
list_of_seqs=[]

test_str = (file2)

matches = re.finditer(regex, test_str, re.MULTILINE)

for matchNum, match in enumerate(matches):
    matchNum = matchNum + 1
    
#     print ("{match}".format(match = match.group()))
    list_of_seqs.append(match.group())
    
    

# Note: for Python 2.7 compatibility, use ur"" to prefix the regex and u"" to prefix the test string and substitution.
```

```python
for i,j in zip(list_of_names,list_of_seqs):
#     print i
#     print j
    file = open(i+".fasta", "w")
    file.write(j) 
    file.close()
```
